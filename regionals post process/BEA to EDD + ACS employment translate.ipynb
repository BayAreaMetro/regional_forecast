```python
import numpy as  np
import pandas as pd
```

```python
#Map the relationship between ACS sectors and REMI sectors.
#We use ACS sector aggregration because self employment data comes from ACS Table C24070, which uses this aggregration
sectormap = pd.read_csv('C:/Users/blu/Box/Modeling and Surveys/Regional Modeling/REMI/control_totals/deliverables/sectormap.csv')
```

```python
sectormap.head()
```

```python
#remi BEA concept employment translates to EDD+ACS concept with a adjustment factor calculated using 2015 data
emp_translate = pd.read_csv('C:/Users/blu/Box/Modeling and Surveys/Regional Modeling/REMI/control_totals/deliverables/emp_translate.csv')
```

```python
emp_translate.head()
```

```python
#read employment data
emp = pd.read_csv('C:/Users/blu/Box/Modeling and Surveys/Regional Modeling/REMI/REMI files/PBA50 Blueprint/REMI output/Forecast 03122020/Blueprint Plus Fix it First/employment_projections.csv')
```

```python
emp_merge = pd.merge(sectormap,emp, how='inner', on=['Industries'])
```

```python
emp_merge
```

```python
emp_sum = emp_merge.groupby(['ind']).sum().reset_index()
```

```python
emp_sum
```

```python
emp_merge2 = pd.merge(emp_sum,emp_translate, how='inner', on=['ind'])
```

```python
emp_merge2.set_index('ind')
```

```python
emp_merge2 [['2017','2020','2025','2030','2035','2040','2045','2050']] = emp_merge2[['2017','2020','2025','2030','2035','2040','2045','2050']].mul(emp_merge2.adjustment, axis = 0)
```

```python
emp_merge2 = emp_merge2.iloc[:, :-1]
```

```python
emp_merge2
```

```python
emp_merge2.to_csv('C:/Users/blu/Box/Modeling and Surveys/Regional Modeling/REMI/REMI files/PBA50 Blueprint/REMI output/Forecast 03122020/Blueprint Plus Fix it First/emp by ind.csv')
```
